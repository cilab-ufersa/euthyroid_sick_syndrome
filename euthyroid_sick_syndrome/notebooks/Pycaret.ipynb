```python
from pycaret.classification import * 
import pandas as pd
```

```python
# abrir o dataset
dataset = pd.read_csv('/content/euthyroid_final_features.csv')
```

```python
dataset = dataset[['classification', 'age', 'sex', 'sick', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']]
```

```python
dataset.head()
```

```python
print(dataset['classification'].value_counts()) #1 = sick, 0 = normal
 
```

```python
# balancear os dados
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42, k_neighbors=5)
dataset_balanced, output_classe = sm.fit_resample(dataset[['age', 'sex', 'sick', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']], dataset['classification'])
```

```python
import pandas as pd 

dataset_b= pd.DataFrame(dataset_balanced)
classes = pd.DataFrame(output_classe)
```

```python
final_dataset = pd.concat([dataset_b, classes], axis=1)
```

```python
final_dataset.head()
```

```python
train_size = 0.8
session = setup(data=final_dataset, target='classification', train_size=train_size, data_split_shuffle=False, normalize = True)
```

```python


best_model = compare_models()

```
